```python
import pandas as pd 
import matplotlib.pyplot as plt 
from datasets import load_dataset


```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
```

```python
#Cleaning
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df
```

```python
paying_countries = df["job_country"].value_counts().head(6).index
hourly_salary = df.pivot_table(
    values="salary_year_avg", 
    index="job_country", 
    columns="job_title_short", 
    aggfunc="max"
)

hourly_salary = hourly_salary.loc[paying_countries]
job_titles = ["Software Engineer", "Data Engineer", "Business Analyst", "Senior Data Analyst"]
hourly_salary = hourly_salary[job_titles]
hourly_salary.plot(kind="bar")
plt.title("Max Salary")
plt.ylabel("Amount in ($USD)")
plt.xlabel("")
plt.xticks(rotation=45, ha="right")
plt.show()

```
