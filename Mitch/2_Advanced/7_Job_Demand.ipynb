```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# #Loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

# #Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df_US = df[df["job_country"] == "United States"].copy()
df_US
```

```python
# Create a new column in df_US dataframe
df_US["job_posted_month"] = df_US["job_posted_date"].dt.strftime("%B") #formats date into a string %B returns the month name eg Jan, feb etc
df_US
```

```python
df_US_pivot = df_US.pivot_table(index="job_posted_month", columns="job_title_short", aggfunc="size")
df_US_pivot.reset_index(inplace=True)
df_US_pivot
```

```python
df_US_pivot["month_no"] = pd.to_datetime(df_US_pivot["job_posted_month"], format="%B").dt.month #converts the month name string into datetime object
# .dt.month- extracts numeric month from the datetime object
df_US_pivot
```

```python
df_US_pivot.sort_values("month_no", inplace=True)
df_US_pivot.set_index("job_posted_month", inplace=True) # sets the column of now "job_posted_month" to became its index ie row label 
df_US_pivot.drop(columns="month_no")

# NB- inplace=True is directly applying change to df instead of creating a new copy
```

```python
df_US_pivot.plot(kind="line")
```

```python
# Pick the top 3 
top_3 = df_US["job_title_short"].value_counts().head(3).index
top_3.to_list()
df_US_pivot[top_3].plot(kind="line")
plt.xlabel("2023")
plt.ylabel("Job Count")
plt.title("Monthly Job Postings for Top Data Jobs in US")
plt.show()
```

```python
last_3 = df_US["job_title_short"].value_counts().tail(3).index.to_list()
df_US_pivot[last_3].plot(kind="line")
plt.xlabel("2023")
plt.ylabel("Job Count")
plt.title("Monthly Job Postings for Least Data Jobs in US")
plt.show()
```
