```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
import ast
import seaborn as sns

# Loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_US = df[(df["job_title"] == "Data Analyst") & (df["job_country"] == "United States")].copy()
df_US
```

```python
df_US["job_posted_month_no"] = df_US["job_posted_date"].dt.month
```

```python
df_US_explode = df_US.explode("job_skills")
df_US_explode
```

```python
df_US_pivot = df_US_explode.pivot_table(index="job_posted_month_no", columns="job_skills", aggfunc="size", fill_value=0)
df_US_pivot.loc["Total"] = df_US_pivot.sum()
df_US_pivot
```

```python
df_US_pivot = df_US_pivot[df_US_pivot.loc["Total"].sort_values(ascending=False).index]
df_US_pivot = df_US_pivot.drop("Total")
df_US_pivot
```

```python
df_US_percentage = df_US.groupby("job_posted_month_no").size()
df_US_percentage
```

```python
df_DA_percent =df_US_pivot.div(df_US_percentage/100, axis=0)
df_DA_percent
```

```python
df_DA_percent = df_DA_percent.reset_index()
df_DA_percent["job_posted_month"] = df_DA_percent["job_posted_month_no"].apply(lambda x: pd.to_datetime(x, format="%m").strftime("%b")) #converts numerical months into jan.feb etc
df_DA_percent = df_DA_percent.set_index("job_posted_month")  # Reassign the result back to df_DA_percent
df_DA_percent = df_DA_percent.drop(columns="job_posted_month_no")  # Reassign the result back to df_DA_percent
df_DA_percent
```

```python
df_plot = df_DA_percent.iloc[:, :5]
df_plot
# Use .loc when you know the labels (row/column names).
# Use .iloc when you want to select by position (like lists).
```

```python
from matplotlib.ticker import PercentFormatter

sns.lineplot(data=df_plot, dashes=False, palette="tab10", legend=False)
sns.set_theme(style="ticks")

plt.title("Trending Top Skills for DA in US")
plt.ylabel("Likelihood in Job Posting")
plt.xlabel("2023")
sns.despine() #used to remove border line except x&y axis
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])

ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))
plt.show()
```
